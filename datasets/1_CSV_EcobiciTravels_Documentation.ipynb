## Ecobici Dataset - Data Source

Este proyecto utiliza el dataset público de afluencia diaria del sistema Ecobici de la Ciudad de México.

### Fuente:
Portal de Datos Abiertos de la CDMX

### Dataset:
https://datos.cdmx.gob.mx/dataset/afluencia-diaria-del-sistema-ecobici

### Recurso (CSV):
https://datos.cdmx.gob.mx/dataset/afluencia-diaria-del-sistema-ecobici/resource/4df66c20-f969-4b3e-9bce-34987da33bc1

### Descripción:
El dataset contiene información diaria del número de viajes realizados en el sistema Ecobici.

### Periodo:
2010 – 2024

### Campos principales:
- fecha
- viajes_totales
- viajes_hombre
- viajes_mujer
- viajes_otro (cuando aplica)

### Formato:
CSV

### Justificación de uso:
Se seleccionó este dataset debido a que, aunque presenta menor nivel de detalle en destinos o trayectorias, ofrece una cobertura histórica amplia y consistente del número de viajes diarios, lo cual lo hace ideal para análisis de tendencias y correlación con variables externas como el clima.

```python
import pandas as pd

url = "https://datos.cdmx.gob.mx/dataset/afluencia-diaria-del-sistema-ecobici/resource/4df66c20-f969-4b3e-9bce-34987da33bc1/download/afluencia_diaria.csv"

df = pd.read_csv(url)

df.head()
```
