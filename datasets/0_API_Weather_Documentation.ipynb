## Weather API - Data Source

Este proyecto utiliza una API pública de clima para obtener información histórica de temperatura y precipitación.

### API utilizada:
Open-Meteo

### Endpoint:
https://archive-api.open-meteo.com/v1/archive

### Parámetros utilizados:
- latitude: ubicación geográfica Ciudad de México
- longitude: ubicación geográfica Ciudad de México
- start_date: fecha inicial
- end_date: fecha final
- daily:
  - temperature_2m_mean
  - precipitation_sum

### Formato de respuesta:
JSON

La información obtenida se utiliza como segunda fuente de datos para el proceso ETL.

```python
import requests

url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": 19.43,
    "longitude": -99.13,
    "start_date": "2024-01-01",
    "end_date": "2024-01-10",
    "daily": "temperature_2m_mean,precipitation_sum",
    "timezone": "auto"
}

response = requests.get(url, params=params)

data = response.json()

print(data.keys())
```

```python
data["daily"]
```

```python
import pandas as pd

df = pd.DataFrame(data["daily"])

df.head()
```
